# Process text

```python
from nltk.corpus import stopwords
import string
from nltk.probability import FreqDist
from collections import Counter
import nltk
WNlemma= nltk.WordNetLemmatizer()
punctuation = list(string.punctuation)+['’','\'','“','”','\'','—','\'','=','/']
#stop_words = stopwords.words('english')+punctuation

## remove punctuation and change the artist, track and album name no matter what format
artist=spotify['artist']
track=spotify['track_name']
album=spotify['album']

def preprocess_text(text,i=1): 
    t=text.lower().replace(artist[i].lower(),"artist")\
    .replace(track[i].lower(),"track")\
    .replace(album[i].lower(),"album")\
    .replace(artist[i].lower().replace(" ",""),"artist")\
    .replace(track[i].lower().replace(" ",""),"track")\
    .replace(album[i].lower().replace(" ",""),"album")
    words=[WNlemma.lemmatize(w) for w in (nltk.word_tokenize(t)) if WNlemma.lemmatize(w) not in punctuation]
    text1=' '.join(words)
    return (text1.strip())
```

```python

text_ready=[]
for i in range(len(text_raw)):
    text=preprocess_text(text_raw[i],1)
    text_ready.append(text)
text_ready[2]

```

## Sentiment Analysis

```python
from textblob import TextBlob
import matplotlib.pyplot as plt

polarity={'Positive':0,'Negative':0}
subjectivity={'Objective':0,'Subjective':0}
for text in text_ready: 
    analysis = TextBlob(text)
    if analysis.sentiment.polarity>0:
        polarity['Positive']+=1
    else:
        polarity['Negative']+=1
    if analysis.sentiment.subjectivity>0.5:
        subjectivity['Subjective']+=1
    else:
        subjectivity['Objective']+=1
print(polarity)
print(polarity['Positive']/(polarity['Positive']+polarity['Negative']))
print(subjectivity)
print(subjectivity['Objective']/(subjectivity['Objective']+subjectivity['Subjective']))
# plt.bar(range(len(polarity)), list(polarity.values()), align='center')
# plt.xticks(range(len(polarity)), list(polarity.keys()))
# plt.show()
# plt.bar(range(len(subjectivity)), list(subjectivity.values()), align='center')
# plt.xticks(range(len(subjectivity)), list(subjectivity.keys()))
# plt.show()
```
